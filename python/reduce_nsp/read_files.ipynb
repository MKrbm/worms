```python
import re
import numpy as np
import pandas as pd
def process_line(stdout):
    energy = []
    sign = []
    time = []
    dimer = []
    weight_r = []
    
    for line in stdout.split("\n"):
        pat = "Total Energy"
        if pat in line:
            line = line.replace(pat,"")
            line = line.replace(" ","")
            line = line.replace("=","")
            line = line.split("+-")
            energy = [float(line[0]), float(line[1])]


        pat = "average sign"
        if pat in line:
            line = line.replace(pat,"")
            line = line.replace(" ","")
            line = line.replace("=","")
            line = line.split("+-")
            sign = [float(line[0]), float(line[1])]

        pat = "Elapsed time"
        if pat in line:
            line = line.replace(pat,"")
            line = line.replace(" ","")
            line = line.replace("=","")
            line = line.replace("sec","")
            time = [float(line)]
            

            
    return energy, sign, time
```

```python
import os
df = {}
folder_path = "make_local/out/SS"
M = 240
N = 1000000
lats = ["dimer_optim", "original", "singlet"]

files = []
Js = np.arange(0.1, 1.2, 0.01)
T =  np.logspace(-1, 0.2, num=20)
for lat in lats:
    df[lat] = {}
    for J in Js:
        df[lat][J] = {}
        for t in T:
            if lat is "dimer_optim":
                file_name = f"{lat}_L_[4,4]_J_[{J:.3},1]_T_{t:.5}_N_{N}_M_{M}"
            else:
                file_name = f"{lat}_L_[4,4]_J_[{J:.3},1]_T_{t:.5}_N_{N}"
            with open(folder_path+ "/" + file_name, "r") as f:
                a = f.read()
                try:
                    energy, sign, time = process_line(a)
                    print(f"J={J}, T={t}, s = {sign[0]}")
                    df[lat][J][t] = sign[0]
                except:
                    print(a)
                    print(file_name)
#                     raise NameError("What?")
                
```

```python
J = 0.3335
Sz = np.zeros([2,2])
Sz[0,0] = 1/2
Sz[1,1] = -1/2
Sx = np.zeros([2,2])
Sx[1,0] = 1/2
Sx[0,1] = 1/2
Sy = np.zeros([2,2], dtype=np.complex64)
Sy[1,0] = 1j/2
Sy[0,1] = -1j/2


SzSz = np.kron(Sz,Sz).real.astype(np.float64)
SxSx = np.kron(Sx,Sx).real.astype(np.float64)
SySy = np.kron(Sy,Sy).real.astype(np.float64)

lh = SzSz + SxSx + SySy

u = np.array([
    [0,1,0,0],
    [1/np.sqrt(2), 0, 1/np.sqrt(2), 0],
    [-1/np.sqrt(2), 0, 1/np.sqrt(2), 0],
    [0,0,0,1]
])

H1 = sum_ham(J*lh, [[1,2],[1,3]], 4, 2)
H1 += sum_ham(lh/4, [[0,1],[2,3]], 4, 2)

H2 = sum_ham(J*lh, [[0,2],[0,3]], 4, 2)
H2 += sum_ham(J*lh/4, [[0,1],[2,3]], 4, 2)
```
